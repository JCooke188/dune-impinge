```python
k   = 0.4
f   = 7.2921*10**-5*2*np.sin(32*np.pi/180)
z0  = 0.01
A0  = 2
B0  = 4.5
G   = 10

def drag(us0):
    global A0, B0, f, z0, k, G
    return k**2/((np.log(us0/f/z0)-A0)**2+B0**2)-us0**2/G**2

us0 = fsolve(drag, G)
print(us0)

#find ug0 somehow then plot a0 (the deflection as a contormap with G and z0 as the axes, 
#will have to find us0 each time by solving the implicit function in a nested loop of G and z0
#this will give you the equilibrium deflection for a steady ABL
# i suppose i really want us and z0 to be the independent variables though...
a0  = np.arctan(f*B0/(np.log(us0/f/z0)-A0))
a0
```
